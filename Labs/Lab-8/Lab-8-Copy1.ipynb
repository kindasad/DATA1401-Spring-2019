** Exercise 4: ** Using numpy, compute the between-class $\bf{S}_B$ and within-class $\bf{S}_W$ covariance matrices defined as:

$$
\bf{S}_B = (\bf{m_2}-\bf{m_1})(\bf{m_2}-\bf{m_1})^T \\
$$
$$
\bf{S}_W = \sum_{i=1,2} \sum_{n=1}^{l_i} (\bf{x}_n^i - \bf{m}_i) (\bf{x}_n^i - \bf{m}_i)^T
$$

where $\bf{m_i}$ are the vectors containing the means for category 1 and 2, here defined as no chest pain and any chest pain. Here $\bf{x}_n^i$ is the vector containing the observables for the $n$th example patient in category $i$.

```python
m_1=list()
m_2=list()
 
for key in df_heart.keys():
    if key!='cp' and key!='target':
        index0=np.where(np_heart[:,h_columns["cp"]]==0.)
        m_1.append(np.mean(np_heart[index0,h_columns[key]].flatten()))
m_1=np.matrix(m_1)

for key in df_heart.keys():
    if key!='cp' and key!='target':
        index1=np.where(np_heart[:,h_columns["cp"]]!=0.)
        m_2.append(np.mean(np_heart[index1,h_columns[key]].flatten()))
m_2=np.matrix(m_2)
 
S_w=np.zeros((12,12))
 
for i in range(303):
    x_1=list()
    x_2=list()

    if df_heart['cp'][i]==0:
        for key in df_heart.keys():
            if key!='cp' and key!='target':
                x_1.append(df_heart[key][i])
    if df_heart['cp'][i]!=0:
        for key in df_heart.keys():
            if key!='cp' and key!='target':
                x_2.append(df_heart[key][i])
    x_1=np.matrix(x_1)
    x_2=np.matrix(x_2)
     
    if (x_1.shape==(1, 0)):
        S_w=S_w+np.transpose(x_2-m_2)* (x_2-m_2)
    if (x_2.shape==(1, 0)):
        S_w=S_w+np.transpose(x_1-m_1)*(x_1-m_1)
 
S_b=np.zeros((12,12))
 
 
S_b=np.transpose(m_2-m_1)* (m_2-m_1)
 
covariance=np.cov(S_b,S_w)
print(covariance.shape)
```

** Exercise 5:** Compute the linear coefficients $\bf{w} = \bf{S_W}^{-1}(\bf{m_2}-\bf{m_1})$. Compare the histogram of the distribution of $F_n^i=\bf{w}^T\bf{x}_n^i$ for the two categories.

```python
w=np.linalg.inv(S_w)
dif=(m_2-m_1)
cof=w*np.transpose(dif)
print(cof.shape)
F_1=0
F_2=0
Fn=list()
F1=list()
F2=list()
#####################
for i in range(303):
    x_1=list()
    x_2=list()

    if df_heart['cp'][i]==0:
        for key in df_heart.keys():
            if key!='cp' and key!='target':
                x_1.append(df_heart[key][i])
    if df_heart['cp'][i]!=0:
        for key in df_heart.keys():
            if key!='cp' and key!='target':
                x_2.append(df_heart[key][i])
    x_1=np.matrix(x_1)
    x_2=np.matrix(x_2)
    if (x_1.shape==(1, 0)):
        F_2=np.transpose(cof)*np.transpose(x_2)+F_2
        F2.append(float(F_2))
        Fn.append(float(F_2))
    if (x_2.shape==(1, 0)):
        F_1=np.transpose(cof)*np.transpose(x_1)+F_1
        F1.append(float(F_1))
        Fn.append(float(F_2))
        
len(Fn)
```
