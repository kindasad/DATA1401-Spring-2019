```python
import math 
import random 
def generate_normal_cos(N):
    out=[]
    while len(out)!=N:
        x=random.random()
        y=random.random()
        a=math.sqrt(math.log(x)*(-2))*math.cos(y*2*math.pi)
        out.append(a)
    return out 
```

```python
print generate_normal_cos(100)
a=generate_normal_cos(100)
```

```python
import math 
import random 
def generate_normal_sin(N):
    out=[]
    while len(out)!=N:
        x=random.random()
        y=random.random()
        a=math.sqrt(math.log(x)*(-2))*math.sin(y*2*math.pi)
        out.append(a)
    return out 

```

```python
print generate_normal_sin(100)
b=generate_normal_sin(100)
```

```python
def mean(x):
    return sum(x)/float(len(x))

def variance(x):
    return sum(map(lambda y:y**2, x))/float(len(x))-mean(x)**2
```

```python
print mean(a)
print variance(a)
print mean(b)
print variance(b)
```
